## Aufgabe a)

```python
fig, ax = plt.subplots()
plt.grid(visible=True, which="both", color="lightgray");
fig.set_dpi(100)

X = np.linspace(-20,20,2000) # X ist der Bereich der X-Koordinaten der geplottet werden soll

standard = np.piecewise(X, [(X >= -3) * (X <= -1)], [1/2, np.NaN]) # Definition der gegebenen Dichtefunktion der ZV.
ax.plot(X,standard, color="green", label="$P_X$") # f_x

p_1 = np.piecewise(X, [(X >= 5) * (X <= 15)], [1/10, np.NaN]) # Definition der Dichte von 1. auf Intervall [5,15]
ax.plot(X,p_1, label="$P_{I}$", color="red") # f_y1

p_2 = np.piecewise(X,[(X >= 0) * (X <= 2)], [1/2, np.NaN]) #Definition der Dichte von 2. auf Intervall [0,2]
ax.plot(X, p_2, label="$P_{II}$")

#bonus - Defintion der Transformation durch x^2 (hier lambda nötig, weil vorher evaluiert)
squared = np.piecewise(X,[(X >= 1) * (X <= 9)], [lambda x: 1/(4*np.sqrt(x)), np.NaN]) #
ax.plot(X, squared, color="blue", label="$P_{III}$", linestyle="--" )

#Setzen der Grenzen des Plots
ax.set_xlim([-5, 20])
ax.set_ylim([0,0.6]);
ax.minorticks_on()

ax.legend();
```

## Aufgabe b)

```python
fig, ax = plt.subplots()
plt.grid(visible=True, which="both", color="lightgray")
fig.set_dpi(100)

#Definition der Normalverteilung als Funktion zum späteren Verwenden
def gaussian(x, sigma, mu):
    return 1./(math.sqrt(np.power(sigma,2)*math.pi*2))*np.exp(-sigma*np.power((x - mu), 2.)/(2*np.power(sigma,2)))

#Definition der gegebenen ZV
X = np.linspace(-20,20,2000)
standard = gaussian(X,4,2)
ax.plot(X,standard, color="green", label="$P_x$") # Plot X

# Transformation 1
first = 1/5*gaussian(((-1/5)*X),4,2)
ax.plot(X, first, color="red", linestyle=":", label="$P_I$") # Plot I.

# Transformation 2
second = gaussian((X-3),4,2)
ax.plot(X, second, color="blue", linestyle="--", label="$P_{II}$") # Plot II.

#Setzen der Grenzen des Plots
ax.set_xlim([-8,8])
ax.set_ylim([0,0.2]);
ax.minorticks_on()

ax.legend();
```
